```python
def juntar_txt(clase):
  """
  Esta Funcion se encarga de recorrer toda la carpeta que se le solicita, levanta los archivos y los va uniendo en uno solo almacenandolos en una variable. 
  """
  tipo = ''
  directorio_deporte = '/content/drive/MyDrive/PNL 2/TP-1/Data/Noticias/' + clase
  with os.scandir(directorio_deporte) as ficheros_deportes:
      for fichero in ficheros_deportes:
          #print(fichero.name)
          
          with open(directorio_deporte + '/' + fichero.name , 'r') as file:
            contenido = file.read()
            tipo += contenido
          
  return tipo
```

```python
entretenimiento = juntar_txt('entretenimiento')
economia = juntar_txt('economia')
deporte = juntar_txt('deporte')
politica = juntar_txt('politica')
```

```python
def final1(text):
  """
  Esta funcion se encarga de hacer todo el preprocesado del texto, pasar a minusculas tokenizar, retirar stop words, luego de eso realiza un conteo de la 
  frecuencia absoluta de las palabras almacenando el resultado en un diccionario con su key = palabra y su value = cantidad de veces dichas. 
  Despues se le aplica una formula para quedarse con la frecuencia realiva de las mismas. 
  """
  tokenizador_expreg = nltk.regexp_tokenize
  texto = text.lower()
  tokens = tokenizador_expreg(re.sub("\d+", "", texto), pattern='\w+')
  tokens = [token.strip() for token in tokens]
  tokens = [token for token in tokens if token.lower() not in stopwords.words("spanish")]

  largo = len(tokens) 
  frab = Counter(tokens)
  frab = dict(frab)

  for key, value in frab.items():
    frab[key] = round(value / largo, 4)

  return frab
```

```python
#Categorias ya listas para ser comparadas
entretenimiento_final = final1(entretenimiento)
economia_final = final1(economia)
deporte_final = final1(deporte)
politica_final = final1(politica)
```

```python
def com_Entretenimiento(entretenimiento_final, nueva_entre): 
  valor = 0
  for x in entretenimiento_final:
    for i in nueva_entre:
      if x == i:
        valor += nueva_entre[i]
  return round(valor, 4)
```

```python
def com_Economia(economia_final, nueva_entre):
  valor = 0
  for x in economia_final:
    for i in nueva_entre:
      if x == i:
        valor += nueva_entre[i]
        #print(f'Es de economia_final {x.upper()}, es de nueva_depo {i.upper()} el valor sumado es {valor}')
  return round(valor, 4)
```

```python
def com_Deporte(deporte_final, nueva_entre):
  valor = 0
  for x in deporte_final:
    for i in nueva_entre:
      if x == i:
        valor += nueva_entre[i]
        #print(f'Es de deporte_final {x.upper()}, es de nueva_depo {i.upper()} el valor sumado es {valor}')
  return round(valor, 4)
```

```python
def com_Politica(politica_final, nueva_entre):  
  valor = 0
  for x in politica_final:

    for i in nueva_entre:
      if x == i:
        valor += nueva_entre[i]
        #print(f'Es de Politica_final {x.upper()}, es de nueva_depo {i.upper()} el valor sumado es {valor}')
  return round(valor, 4)
```

```python
# Comparacion Entretenimiento.
valor_Entretenimiento = com_Entretenimiento(entretenimiento_final, nueva_entre)
valor_Economia = com_Economia(economia_final, nueva_entre)
valor_Deporte = com_Deporte(deporte_final, nueva_entre)
valor_Politica = com_Politica(politica_final, nueva_entre)

print(f"El valor de Entretenimiento es: {valor_Entretenimiento}")
print(f"El valor de Economia es: {valor_Economia}")
print(f"El valor de Deporte es: {valor_Deporte}")
print(f"El valor de Politica es: {valor_Politica}")
```
